```python
def read_fasta(filename):
    seqs = list()
    pdb_ids = list()
    seq = ''
    pdb_id = ''
    with open(filename, 'r') as f:
        for line in f:
            line = line.strip()
            if line.startswith('>'):
                if seq != '':
                    seqs.append(seq)
                    pdb_ids.append(pdb_id)
                    seq = ''
                inf = line[1:].split('|')[0]
                pdb_id = inf.lower()
            else:
                seq += line
        seqs.append(seq)
        pdb_ids.append(pdb_id)
    return pdb_ids, seqs
```

```python
pdb_ids, seqs = read_fasta('rcsb_pdb_20220816204656.fasta')
```

```python
len(set(pdb_ids)),len(seqs)
```

```python
pid_list = []
seq_list = []
for pid,seq in zip(pdb_ids,seqs):
    if pid.endswith('1'):
        pid = pid.split('_')[0]
        if pid not in pid_list:
            pid_list.append(pid)
            seq_list.append(seq)
```

```python
len(set(pid_list)),len(set(seq_list))
```

```python
df_seq = pd.DataFrame(data={'pdb_id':pid_list,'sequence':seq_list})
```
