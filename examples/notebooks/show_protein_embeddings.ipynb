```python
import os
import sys
import pandas as pd
import numpy as np
sys.path.append('../../')
from deepfold.data.utils.ontology import Ontology
from deepfold.utils.file_utils import read_embeddings
```

```python
base_path = '/home/niejianzheng/xbiome/datasets/protein'
# 存储所有数据文件路径
data_ls = os.walk(base_path).__next__()[2]
data_path_dict = {}
for data in data_ls:
    file_name = data.split('.')[0] + '_' + data.split('.')[1]
    data_path_dict[file_name] = os.path.join(base_path, data)
```

```python
data_path_dict
```

```python
go_rels = Ontology(data_path_dict['go_obo'], with_rels=True)

test_df = pd.read_pickle(data_path_dict['esm1b_t33_650M_UR50S_embeddings_mean_test_pkl'])
test_annotations = test_df['prop_annotations'].values
test_annotations = list(map(lambda x: set(x), test_annotations))
```

```python
def get_species(x):
    sp = x.split('_')[-1]
    return sp


test_df['species'] = test_df['proteins'].map(get_species)
```

```python
test_df.head()
```

```python
from sklearn.manifold import TSNE
import plotly.express as px
from sklearn.preprocessing import StandardScaler


features = np.array(test_df['esm_embeddings'].tolist())

tsne = TSNE(n_components=2, random_state=0)
projections = tsne.fit_transform(features)
```

```python
fig = px.scatter(
    projections,
    x=0, y=1,
    color=test_df.species,
    labels={'color': 'species'},
    width=1200, height=800
)
fig.show()
```

```python
import umap
reducer = umap.UMAP()
embedding = reducer.fit_transform(features)


fig = px.scatter(
    embedding,
    x=0, y=1,
    color=test_df.species,
    labels={'color': 'species'},
    width=1200, height=800
)
fig.show()
```

```python
features = np.array(test_df['esm_embeddings'].tolist())
features = StandardScaler().fit_transform(features.transpose()).transpose()

tsne = TSNE(n_components=2, random_state=0)
projections = tsne.fit_transform(features)
```

```python
fig = px.scatter(
    projections,
    x=0, y=1,
    color=test_df.species,
    labels={'color': 'species'},
    width=1200, height=800
)
fig.show()
```

```python
import umap
reducer = umap.UMAP()
embedding = reducer.fit_transform(features)

fig = px.scatter(
    embedding,
    x=0, y=1,
    color=test_df.species,
    labels={'color': 'species'},
    width=1200, height=800
)
fig.show()
```

```python
embedding = umap.UMAP(n_neighbors=3).fit_transform(features)


fig = px.scatter(
    embedding,
    x=0, y=1,
    color=test_df.species,
    labels={'color': 'species'},
    width=1200, height=800
)
fig.show()
```
